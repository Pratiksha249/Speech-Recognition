```python
import numpy as np
import matplotlib.pyplot as plt

# --- Parameters ---
fs = 1000  # Sampling frequency in Hz
f = 5      # Signal frequency in Hz
duration = 1 # Duration in seconds
N = int(fs * duration) # Total number of samples
t = np.linspace(0, duration, N, endpoint=False) # Time vector

# (a) Generate the sinusoidal signal
x_sin = np.sin(2 * np.pi * f * t)

# (b) Plot the time-domain waveform
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
plt.plot(t, x_sin)
plt.title('Time-Domain: 5 Hz Sinusoidal Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.xlim(0, 0.5) # Zoom in to see the wave shape clearly
plt.tight_layout()
plt.show()
```

```python
# (c) Compute and Plot the DTFT (approximated with a high-resolution DFT)
# We use zero-padding to get a smoother, continuous-looking curve.
N_dtft = 4096 # Number of points for a high-resolution transform
X_dtft = np.fft.fft(x_sin, n=N_dtft)
freq_dtft = np.fft.fftfreq(N_dtft, d=1/fs)

# We only plot the positive frequencies for clarity (0 to fs/2)
positive_freq_indices_dtft = np.where(freq_dtft >= 0)

plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
plt.plot(freq_dtft[positive_freq_indices_dtft], np.abs(X_dtft[positive_freq_indices_dtft]))
plt.title('Frequency-Domain: DTFT (Continuous Spectrum)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.xlim(0, 50) # Zoom in around the 5 Hz peak

# (d) Compute and Plot the DFT (Discrete Spectrum)
X_dft = np.fft.fft(x_sin)
freq_dft = np.fft.fftfreq(N, d=1/fs)

# We only plot the positive frequencies
positive_freq_indices_dft = np.where(freq_dft >= 0)

plt.subplot(2, 1, 2)
plt.stem(freq_dft[positive_freq_indices_dft], np.abs(X_dft[positive_freq_indices_dft]))
plt.title('Frequency-Domain: DFT (Discrete Spectrum)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.xlim(0, 50) # Zoom in around the 5 Hz peak
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Parameters ---
fs = 1000
duration = 1
N = int(fs * duration)
t = np.linspace(0, duration, N, endpoint=False)

# (a) Generate the composite signal (15 Hz and 40 Hz waves)
x1 = 1.0 * np.sin(2 * np.pi * 15 * t) # 15 Hz wave with amplitude 1.0
x2 = 0.5 * np.sin(2 * np.pi * 40 * t) # 40 Hz wave with amplitude 0.5
x_composite = x1 + x2

# (b) Plot the time-domain waveform
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
plt.plot(t, x_composite)
plt.title('Time-Domain: Composite Signal (15 Hz + 40 Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.xlim(0, 0.2) # Zoom in to see the complex shape
plt.tight_layout()
plt.show()
```

```python
# (c) Compute and Plot the DTFT
N_dtft = 4096
X_dtft_comp = np.fft.fft(x_composite, n=N_dtft)
freq_dtft = np.fft.fftfreq(N_dtft, d=1/fs)
positive_freq_indices_dtft = np.where(freq_dtft >= 0)

plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
plt.plot(freq_dtft[positive_freq_indices_dtft], np.abs(X_dtft_comp[positive_freq_indices_dtft]))
plt.title('Frequency-Domain: DTFT of Composite Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.xlim(0, 100)

# (d) Compute and Plot the DFT
X_dft_comp = np.fft.fft(x_composite)
freq_dft = np.fft.fftfreq(N, d=1/fs)
positive_freq_indices_dft = np.where(freq_dft >= 0)

plt.subplot(2, 1, 2)
plt.stem(freq_dft[positive_freq_indices_dft], np.abs(X_dft_comp[positive_freq_indices_dft]))
plt.title('Frequency-Domain: DFT of Composite Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.xlim(0, 100)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Parameters ---
fs = 1000
duration = 2
N = int(fs * duration)
t = np.linspace(0, duration, N, endpoint=False)
decay_rate = 2.0 # Controls how fast the signal decays

# (a) Generate the exponentially decaying signal
x_exp = np.exp(-decay_rate * t)

# (b) Plot the time-domain waveform
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
plt.plot(t, x_exp)
plt.title('Time-Domain: Exponentially Decaying Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# (c) & (d) Compute and Plot DTFT and DFT
N_dtft = 4096
X_dtft_exp = np.fft.fft(x_exp, n=N_dtft)
freq_dtft = np.fft.fftfreq(N_dtft, d=1/fs)
positive_freq_indices_dtft = np.where(freq_dtft >= 0)

plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
plt.plot(freq_dtft[positive_freq_indices_dtft], np.abs(X_dtft_exp[positive_freq_indices_dtft]))
plt.title('Frequency-Domain: DTFT of Exponential Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.xlim(0, 20)

X_dft_exp = np.fft.fft(x_exp)
freq_dft = np.fft.fftfreq(N, d=1/fs)
positive_freq_indices_dft = np.where(freq_dft >= 0)

plt.subplot(2, 1, 2)
plt.plot(freq_dft[positive_freq_indices_dft], np.abs(X_dft_exp[positive_freq_indices_dft]), 'o-')
plt.title('Frequency-Domain: DFT of Exponential Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.xlim(0, 20)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Parameters ---
fs = 1000
duration = 1
N = int(fs * duration)
t = np.linspace(0, duration, N, endpoint=False)

# (a) Generate a rectangular pulse
x_rect = np.zeros(N)
# Make the pulse 'on' from t=0.4s to t=0.6s
pulse_start_index = int(0.4 * fs)
pulse_end_index = int(0.6 * fs)
x_rect[pulse_start_index:pulse_end_index] = 1

# (b) Plot the time-domain waveform
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
plt.plot(t, x_rect)
plt.title('Time-Domain: Rectangular Pulse')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.ylim(-0.1, 1.1)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# (c) & (d) Compute and Plot DTFT and DFT
N_dtft = 4096
X_dtft_rect = np.fft.fft(x_rect, n=N_dtft)
freq_dtft = np.fft.fftfreq(N_dtft, d=1/fs)

# We need to use fftshift to center the spectrum at 0 Hz for plotting the Sinc shape
X_dtft_rect_shifted = np.fft.fftshift(X_dtft_rect)
freq_dtft_shifted = np.fft.fftshift(freq_dtft)

plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
plt.plot(freq_dtft_shifted, np.abs(X_dtft_rect_shifted))
plt.title('Frequency-Domain: DTFT of Rectangular Pulse (Sinc Function)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.xlim(-50, 50)

X_dft_rect = np.fft.fft(x_rect)
freq_dft = np.fft.fftfreq(N, d=1/fs)
X_dft_rect_shifted = np.fft.fftshift(X_dft_rect)
freq_dft_shifted = np.fft.fftshift(freq_dft)

plt.subplot(2, 1, 2)
plt.plot(freq_dft_shifted, np.abs(X_dft_rect_shifted), 'o-')
plt.title('Frequency-Domain: DFT of Rectangular Pulse')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.xlim(-50, 50)
plt.tight_layout()
plt.show()
```
